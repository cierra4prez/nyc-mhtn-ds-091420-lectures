Matplotlib is a complex library (70,000 + lines of code!). In order to best understand how it creates graphs it's important to understand how objects are structured in matplotlib. 

### Two Ways to Generate a Single Plot

**Method 1**<br>
Plot your Xs vs your Ys

```python
##not the best - not storing the image, just looking at the data here. can't customize much (use )

X = [1,2]
Y = [3,4]
plt.scatter(X,Y); #calls scatter inside library
plt.title('This is an example title')
```

**Method 2**<br>
Using subplots to set up your figure and axes

<img src = 'figure-axes.png' width = 400;>

```python
#figure - whole picture (cannot be changed, just a display of an image)
#axes - customizable

fig, ax = plt.subplots() #creating 2 objects: figure and ax
ax.scatter(X, Y); #call plot method associate with ...
```

```python
type(fig) #tells me it's just an object
type(ax) #tells me ...
```

```python
ax.set_title('This is an example title'); #title
ax.set_xlabel('x label') #x axis
ax.set_ylabel('i made a label')
ax.set_xticks([1,2]) #how many ticks you want on x
ax.set_xticklabels(['one','dog']) #what each tick stands for
fig
# examine the ax object
```

```python
# examine the figure
```

#### How do we add a title to the above plot? 

```python
ax.set_title('Line')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis');
ax.set_xticks([1,2]);
```

```python
# examine the figure
fig
```

### Another example

```python
#generate some data
x = np.linspace(0, 10, 100) #start at 1, end at 10, across 100 data points
y = np.sin(x)
```

#### create a figure with 2 x 2 plots

```python
figure_sin, axes = plt.subplots(2,2, figsize = (10,10)) #creating 2 objects
```

```python
axes[0][0].scatter(x,y) #first row first column, scatter
axes[0][1].plot(x,y) #first row second column, line
axes[1][0].hist(y) #histogram
axes[1][1].fill(x,y); #fill plot
```

```python
figure_sin
```

#### How can we create a title for the scatter plot? 

```python
axes[0][0].set_title('Scatter')
axes[0][0].set_xlabel('Independent variable name')
axes[0][0].set_ylabel('Dependent variable name');
axes[0][1].set_title('line plot');
```

```python
figure_sin
```

## Data Analysis Example & Using Seaborn

Seaborn is built on top of Matplotlib. Seaborn adds some styling to existing matplotlib graphs as well as adds some additional graphing capabilities. 

```python
import seaborn as sns
sns.set()
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.plot([1,2], [3,4])
```

The most useful aspect of seaborn is that it allows you to use Pandas DataFrame Objects directly. 

```python
#loads tips dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# visualize the relationship between bill and tip
sns.scatterplot('total_bill', 'tip', data=tips);
```

```python
tips.sex
```

```python
# calculate average tips amount by gender
#use pandas and groupby

y = tips.groupby(['sex']).tip.mean() #tip
x = y.index #F/M
```

```python
# check to see if x and y are correct
print(y)
print(y.index)
```

```python
# visualize the average tips men vs women pay - what's the appropriate plot?
##bar plot
#x = None
#average_tips_amount = None
fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_title('average tip amount for male and female')
ax.set_xticks([0,1])
ax.set_xticklabels(['male','female'])
```

```python
# visualize the average amount of total bill for time (lunch or dinner)
y = None
x = None
fig, ax = plt.subplots()
ax.bar(x,y, color = 'pink')
ax.set_title('average tip amount for time')
ax.set_xticks([0,1])
ax.set_xticklabels(['lunch','dinner'])
```
